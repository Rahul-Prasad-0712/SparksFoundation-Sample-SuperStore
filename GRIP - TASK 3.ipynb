### Step 6:- Data Visualization

```python
# Creating A BAr Plot In Sales And Category
plt.bar(df['Category'],df['Sales'])
plt.xlabel('Category')
plt.ylabel('Sales')
```

#### Here by plotting this bar plot we get to know that 'Sales' are high if the Category is 'Technology', followed by 'Office Supplies' and 'Furniture'.

```python
# Creating A BAr Plot In Ship Mode And Sales
plt.bar(df['Ship Mode'],df['Sales'])
plt.xlabel('Ship Mode')
plt.ylabel('Sales')
```

#### Here by plotting this bar plot we get to know that 'Sales' are high if the Ship Mode is 'Standard Class' and then followed by 'First Class', 'Second Class', 'Same Day'

```python
sns.countplot(x=df.Segment)
plt.show();
     
```

#### Here in this Countplot We observe that 'Consumer' has the most number of count followed by 'Corporate Segment' and 'Home Office'

```python
sns.countplot(x='Category',data=df)
plt.show()
```

#### Here in this Countplot We observe that 'Office Supplies Category' has the most number of count followed by 'Furniture' and 'Technology'
